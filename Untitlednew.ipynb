```python
import warnings

# To ignore all warnings
warnings.filterwarnings("ignore")

import pandas as pd
import datetime
import numpy as np
TF=0.07/100
regime_niftyd=25 #Bank_nav
regime_bnd=25#Nifty_nav
regime_niftyh=25#Bank_nav
regime_bnh=25 #Nifty_nav
#vol_factor_average=30*24
rebalance_days=30


sig_df=pd.read_csv("NIFTY 50h_result.csv",index_col="Date",parse_dates=True)
df_add=pd.read_csv("NIFTY BANKh_result.csv",index_col="Date",parse_dates=True)
df_add1=pd.read_csv("NIFTY 50d_result.csv",index_col="Date",parse_dates=True)
df_add2=pd.read_csv("NIFTY BANKd_result.csv",index_col="Date",parse_dates=True)

sig_df["NIFTY BANKh_Close"]=df_add["Close"]
sig_df["NIFTY BANKh_Signal"]=df_add["Signal"]
sig_df["NIFTY BANKh_Price"]=df_add["Price"]
sig_df["NIFTY BANKh_Price_exit"]=df_add["Price_exit"]

sig_df["NIFTY 50d_Close"]=df_add1["Close"]
sig_df["NIFTY 50d_Signal"]=df_add1["Signal"]
sig_df["NIFTY 50d_Price"]=df_add1["Price"]
sig_df["NIFTY 50d_Price_exit"]=df_add1["Price_exit"]

sig_df["NIFTY BANKd_Close"]=df_add2["Close"]
sig_df["NIFTY BANKd_Signal"]=df_add2["Signal"]
sig_df["NIFTY BANKd_Price"]=df_add2["Price"]
sig_df["NIFTY BANKd_Price_exit"]=df_add2["Price_exit"]

sig_df["NIFTY 50h_pctcg"]=0.0
sig_df["NIFTY BANKh_pctcg"]=0.0
sig_df["NIFTY 50d_pctcg"]=0.0
sig_df["NIFTY BANKd_pctcg"]=0.0



sig_df=sig_df.rename(columns={'Close':'NIFTY 50h_Close','Signal': 'NIFTY 50h_Signal', 'Price': 'NIFTY 50h_Price','Price_exit':'NIFTY 50h_Price_exit'})
#sig_df["date"]=sig_df.index.date



sig_df["NIFTY 50h_Price"]=np.where((sig_df["NIFTY 50h_Signal"]==sig_df["NIFTY 50h_Signal"].shift()) & (sig_df["NIFTY 50h_Signal"]!=0),
                             sig_df["NIFTY 50h_Close"],
                             sig_df["NIFTY 50h_Price"]          
                             )

sig_df["NIFTY 50h_Price"]=np.where((sig_df["NIFTY 50h_Signal"]!=sig_df["NIFTY 50h_Signal"].shift()) & (sig_df["NIFTY 50h_Signal"]!=0),
                             sig_df["NIFTY 50h_Price"]+(sig_df["NIFTY 50h_Price"]*TF*sig_df["NIFTY 50h_Signal"]),
                             sig_df["NIFTY 50h_Price"]          
                             )

sig_df["NIFTY 50h_Price"]=np.where((sig_df["NIFTY 50h_Signal"]!=sig_df["NIFTY 50h_Signal"].shift()) & (sig_df["NIFTY 50h_Signal"]==0),
                             sig_df["NIFTY 50h_Price"]-(sig_df["NIFTY 50h_Price"]*TF*sig_df["NIFTY 50h_Signal"].shift()),
                             sig_df["NIFTY 50h_Price"]          
                             )

sig_df["NIFTY 50h_Price_exit"]=np.where((sig_df["NIFTY 50h_Signal"]!=sig_df["NIFTY 50h_Signal"].shift()) & (sig_df["NIFTY 50h_Signal"]!=0) & (sig_df["NIFTY 50h_Signal"].shift()!=0),
                             sig_df["NIFTY 50h_Price_exit"]-(sig_df["NIFTY 50h_Price_exit"]*TF*sig_df["NIFTY 50h_Signal"].shift()),
                             sig_df["NIFTY 50h_Price_exit"]          
                             )


sig_df["NIFTY 50h_pctcg"]=np.where((sig_df["NIFTY 50h_Signal"]==sig_df["NIFTY 50h_Signal"].shift()) & (sig_df["NIFTY 50h_Signal"]!=0),
                             sig_df["NIFTY 50h_Price"].pct_change()*sig_df["NIFTY 50h_Signal"],
                             sig_df["NIFTY 50h_pctcg"]          
                             )

sig_df["NIFTY 50h_pctcg"]=np.where((sig_df["NIFTY 50h_Signal"]!=sig_df["NIFTY 50h_Signal"].shift()) & (sig_df["NIFTY 50h_Signal"]==0),
                             sig_df["NIFTY 50h_Price"].pct_change()*sig_df["NIFTY 50h_Signal"].shift(),
                             sig_df["NIFTY 50h_pctcg"]          
                             )

sig_df["NIFTY 50h_pctcg"]=np.where((sig_df["NIFTY 50h_Signal"]!=sig_df["NIFTY 50h_Signal"].shift()) & (sig_df["NIFTY 50h_Signal"]!=0) & (sig_df["NIFTY 50h_Signal"].shift()!=0),
                             ((sig_df["NIFTY 50h_Price_exit"]-sig_df["NIFTY 50h_Price"].shift())/sig_df["NIFTY 50h_Price"].shift())*sig_df["NIFTY 50h_Signal"].shift(),
                             sig_df["NIFTY 50h_pctcg"]          
                             )
    

sig_df["NIFTY 50h_Invested"]=np.where(sig_df["NIFTY 50h_Signal"]!=0,1,0)


sig_df["NIFTY BANKh_Price"]=np.where((sig_df["NIFTY BANKh_Signal"]==sig_df["NIFTY BANKh_Signal"].shift()) & (sig_df["NIFTY BANKh_Signal"]!=0),
                             sig_df["NIFTY BANKh_Close"],
                             sig_df["NIFTY BANKh_Price"]          
                             )

sig_df["NIFTY BANKh_Price"]=np.where((sig_df["NIFTY BANKh_Signal"]!=sig_df["NIFTY BANKh_Signal"].shift()) & (sig_df["NIFTY BANKh_Signal"]!=0),
                             sig_df["NIFTY BANKh_Price"]+(sig_df["NIFTY BANKh_Price"]*TF*sig_df["NIFTY BANKh_Signal"]),
                             sig_df["NIFTY BANKh_Price"]          
                             )

sig_df["NIFTY BANKh_Price"]=np.where((sig_df["NIFTY BANKh_Signal"]!=sig_df["NIFTY BANKh_Signal"].shift()) & (sig_df["NIFTY BANKh_Signal"]==0),
                             sig_df["NIFTY BANKh_Price"]-(sig_df["NIFTY BANKh_Price"]*TF*sig_df["NIFTY BANKh_Signal"].shift()),
                             sig_df["NIFTY BANKh_Price"]          
                             )

sig_df["NIFTY BANKh_Price_exit"]=np.where((sig_df["NIFTY BANKh_Signal"]!=sig_df["NIFTY BANKh_Signal"].shift()) & (sig_df["NIFTY BANKh_Signal"]!=0) & (sig_df["NIFTY BANKh_Signal"].shift()!=0),
                             sig_df["NIFTY BANKh_Price_exit"]-(sig_df["NIFTY BANKh_Price_exit"]*TF*sig_df["NIFTY BANKh_Signal"].shift()),
                             sig_df["NIFTY BANKh_Price_exit"]          
                             )


sig_df["NIFTY BANKh_pctcg"]=np.where((sig_df["NIFTY BANKh_Signal"]==sig_df["NIFTY BANKh_Signal"].shift()) & (sig_df["NIFTY BANKh_Signal"]!=0),
                             sig_df["NIFTY BANKh_Price"].pct_change()*sig_df["NIFTY BANKh_Signal"],
                             sig_df["NIFTY BANKh_pctcg"]          
                             )

sig_df["NIFTY BANKh_pctcg"]=np.where((sig_df["NIFTY BANKh_Signal"]!=sig_df["NIFTY BANKh_Signal"].shift()) & (sig_df["NIFTY BANKh_Signal"]==0),
                             sig_df["NIFTY BANKh_Price"].pct_change()*sig_df["NIFTY BANKh_Signal"].shift(),
                             sig_df["NIFTY BANKh_pctcg"]          
                             )

sig_df["NIFTY BANKh_pctcg"]=np.where((sig_df["NIFTY BANKh_Signal"]!=sig_df["NIFTY BANKh_Signal"].shift()) & (sig_df["NIFTY BANKh_Signal"]!=0) & (sig_df["NIFTY BANKh_Signal"].shift()!=0),
                             ((sig_df["NIFTY BANKh_Price_exit"]-sig_df["NIFTY BANKh_Price"].shift())/sig_df["NIFTY BANKh_Price"].shift())*sig_df["NIFTY BANKh_Signal"].shift(),
                             sig_df["NIFTY BANKh_pctcg"]          
                             )
    

sig_df["NIFTY BANKh_Invested"]=np.where(sig_df["NIFTY BANKh_Signal"]!=0,1,0)


sig_df["NIFTY 50d_Price"]=np.where((sig_df["NIFTY 50d_Signal"]==sig_df["NIFTY 50d_Signal"].shift()) & (sig_df["NIFTY 50d_Signal"]!=0),
                             sig_df["NIFTY 50d_Close"],
                             sig_df["NIFTY 50d_Price"]          
                             )

sig_df["NIFTY 50d_Price"]=np.where((sig_df["NIFTY 50d_Signal"]!=sig_df["NIFTY 50d_Signal"].shift()) & (sig_df["NIFTY 50d_Signal"]!=0),
                             sig_df["NIFTY 50d_Price"]+(sig_df["NIFTY 50d_Price"]*TF*sig_df["NIFTY 50d_Signal"]),
                             sig_df["NIFTY 50d_Price"]          
                             )

sig_df["NIFTY 50d_Price"]=np.where((sig_df["NIFTY 50d_Signal"]!=sig_df["NIFTY 50d_Signal"].shift()) & (sig_df["NIFTY 50d_Signal"]==0),
                             sig_df["NIFTY 50d_Price"]-(sig_df["NIFTY 50d_Price"]*TF*sig_df["NIFTY 50d_Signal"].shift()),
                             sig_df["NIFTY 50d_Price"]          
                             )

sig_df["NIFTY 50d_Price_exit"]=np.where((sig_df["NIFTY 50d_Signal"]!=sig_df["NIFTY 50d_Signal"].shift()) & (sig_df["NIFTY 50d_Signal"]!=0) & (sig_df["NIFTY 50d_Signal"].shift()!=0),
                             sig_df["NIFTY 50d_Price_exit"]-(sig_df["NIFTY 50d_Price_exit"]*TF*sig_df["NIFTY 50d_Signal"].shift()),
                             sig_df["NIFTY 50d_Price_exit"]          
                             )


sig_df["NIFTY 50d_pctcg"]=np.where((sig_df["NIFTY 50d_Signal"]==sig_df["NIFTY 50d_Signal"].shift()) & (sig_df["NIFTY 50d_Signal"]!=0),
                             sig_df["NIFTY 50d_Price"].pct_change()*sig_df["NIFTY 50d_Signal"],
                             sig_df["NIFTY 50d_pctcg"]          
                             )

sig_df["NIFTY 50d_pctcg"]=np.where((sig_df["NIFTY 50d_Signal"]!=sig_df["NIFTY 50d_Signal"].shift()) & (sig_df["NIFTY 50d_Signal"]==0),
                             sig_df["NIFTY 50d_Price"].pct_change()*sig_df["NIFTY 50d_Signal"].shift(),
                             sig_df["NIFTY 50d_pctcg"]          
                             )

sig_df["NIFTY 50d_pctcg"]=np.where((sig_df["NIFTY 50d_Signal"]!=sig_df["NIFTY 50d_Signal"].shift()) & (sig_df["NIFTY 50h_Signal"]!=0) & (sig_df["NIFTY 50d_Signal"].shift()!=0),
                             sig_df["NIFTY 50d_Close"].pct_change(),
                             sig_df["NIFTY 50d_pctcg"]          
                             )
    

sig_df["NIFTY 50d_Invested"]=np.where(sig_df["NIFTY 50d_Signal"]!=0,1,0)


sig_df["NIFTY BANKd_Price"]=np.where((sig_df["NIFTY BANKd_Signal"]==sig_df["NIFTY BANKd_Signal"].shift()) & (sig_df["NIFTY BANKd_Signal"]!=0),
                             sig_df["NIFTY BANKd_Close"],
                             sig_df["NIFTY BANKd_Price"]          
                             )

sig_df["NIFTY BANKd_Price"]=np.where((sig_df["NIFTY BANKd_Signal"]!=sig_df["NIFTY BANKd_Signal"].shift()) & (sig_df["NIFTY BANKd_Signal"]!=0),
                             sig_df["NIFTY BANKd_Price"]+(sig_df["NIFTY BANKd_Price"]*TF*sig_df["NIFTY BANKd_Signal"]),
                             sig_df["NIFTY BANKd_Price"]          
                             )

sig_df["NIFTY BANKd_Price"]=np.where((sig_df["NIFTY BANKd_Signal"]!=sig_df["NIFTY BANKd_Signal"].shift()) & (sig_df["NIFTY BANKd_Signal"]==0),
                             sig_df["NIFTY BANKd_Price"]-(sig_df["NIFTY BANKd_Price"]*TF*sig_df["NIFTY BANKd_Signal"].shift()),
                             sig_df["NIFTY BANKd_Price"]          
                             )

sig_df["NIFTY BANKd_Price_exit"]=np.where((sig_df["NIFTY BANKd_Signal"]!=sig_df["NIFTY BANKd_Signal"].shift()) & (sig_df["NIFTY BANKd_Signal"]!=0) & (sig_df["NIFTY BANKd_Signal"].shift()!=0),
                             sig_df["NIFTY BANKd_Price_exit"]-(sig_df["NIFTY BANKd_Price_exit"]*TF*sig_df["NIFTY BANKd_Signal"].shift()),
                             sig_df["NIFTY BANKd_Price_exit"]          
                             )


sig_df["NIFTY BANKd_pctcg"]=np.where((sig_df["NIFTY BANKd_Signal"]==sig_df["NIFTY BANKd_Signal"].shift()) & (sig_df["NIFTY BANKd_Signal"]!=0),
                             sig_df["NIFTY BANKd_Price"].pct_change()*sig_df["NIFTY BANKd_Signal"],
                             sig_df["NIFTY BANKd_pctcg"]          
                             )

sig_df["NIFTY BANKd_pctcg"]=np.where((sig_df["NIFTY BANKd_Signal"]!=sig_df["NIFTY BANKd_Signal"].shift()) & (sig_df["NIFTY BANKd_Signal"]==0),
                             sig_df["NIFTY BANKd_Price"].pct_change()*sig_df["NIFTY BANKd_Signal"].shift(),
                             sig_df["NIFTY BANKd_pctcg"]          
                             )

sig_df["NIFTY BANKd_pctcg"]=np.where((sig_df["NIFTY BANKd_Signal"]!=sig_df["NIFTY BANKd_Signal"].shift()) & (sig_df["NIFTY BANKd_Signal"]!=0) & (sig_df["NIFTY BANKd_Signal"].shift()!=0),
                             sig_df["NIFTY BANKd_Close"].pct_change(),
                             sig_df["NIFTY BANKd_pctcg"]          
                             )
    

sig_df["NIFTY BANKd_Invested"]=np.where(sig_df["NIFTY BANKd_Signal"]!=0,1,0)



sig_df["Total_Invested"]=sig_df["NIFTY BANKh_Invested"]+sig_df["NIFTY 50h_Invested"]+sig_df["NIFTY BANKd_Invested"]+sig_df["NIFTY 50d_Invested"]
```
